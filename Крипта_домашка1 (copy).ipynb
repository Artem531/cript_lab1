```python
import numpy as np

#text_stream = open("Крипта_текст.txt")

text = ["a","r","t","e","m"]

#for string in text_stream:
#  text += [c for c in string.lower()]
  
print(text)

key = "key"

alphabet = np.array(list('abcdefghijklmnopqrstuvwxyz'))
first_string = list('abcdefghijklmnopqrstuvwxyz')
print(first_string)

def gen_table(first_string):
    """Generate a Vigenere table"""
    result = []
    for i in range(len(first_string)):
        result.append(first_string[i:] + first_string[:i])
    return result

table = np.array(gen_table(first_string))


def encode(table, text, key):
    new_text = []
    for idx, c in enumerate (text):
        key_idx = idx % len(key)
        idx1 = np.where(alphabet == c)[0]
        idx2 = np.where(alphabet == key[key_idx])[0]
        if idx1 != None:
            new_text += [table[idx1, idx2]]
        else:
            new_text += [c]
      
    return new_text

cript_new_text = encode(table, text, key)
cript_new_text = np.array(cript_new_text)
cript_new_text = cript_new_text.reshape(-1)
print(cript_new_text)

print(len(cript_new_text))
print(len(text))
assert len(cript_new_text) == len(text), 'len must be the same'

def decode(table, text, key):
    new_text = []
    for idx, c in enumerate (text):
        key_idx = idx % len(key)
        idx2 = np.where(alphabet == key[key_idx])[0]
        idx1 = np.where(table[:, idx2] == c)[0]
        if idx1 != None and idx2 != None:
            new_text.append(alphabet[idx1])
        else:
            new_text.append(c)
    return new_text

decode_new_text = decode(table, cript_new_text, key)
decode_new_text = np.array(decode_new_text).reshape(-1)
print(decode_new_text)
```

```python
import numpy as np

# get text and alphabet
text = ["a","r","t","e","m", '%']
alphabet = np.array(list('abcdefghijklmnopqrstuvwxyz ,.:;-!?123456789') + ['\n'])
key = ['k','e','y']

# get ord lists
ord_alphabet = list(map(lambda i: i, range(len(alphabet) )))

def get_ord(text, alphabet):
    ord_text = list(map(lambda i: -1 if  len(np.where(i == alphabet)[0]) == 0 else np.where(i == alphabet)[0][0], text ))
    print("len ord_text", len(ord_text))
    print("check",  np.sum(ord_text))
    tmp = []
    for i in ord_text:
        if i == -1:
            continue
        tmp.append(int(i))
    ord_text = np.array(tmp)
    return ord_text

ord_text = get_ord(text, alphabet)
key = get_ord(key, alphabet)


print("key, ord_text, ord_alphabet", key, ord_text, ord_alphabet)


def encode(text, alphabet, key):
    new_text = text
    print(text)
    print("real len of key", len(key))
    for i in range(0, len(text)):
        key_idx = i % len(key)
        #print(new_text[i], key[key_idx])
        new_text[i] = (new_text[i] + key[key_idx]) % len(alphabet) 
    print("encode", new_text)
    return new_text


def decode(text, alphabet, key):
    new_text = text
    for i in range(0, len(text)):
        key_idx = i % len(key)
        #print(new_text[i], key[key_idx])
        new_text[i] = (new_text[i] - key[key_idx]) % len(alphabet)     
    return new_text

# encode
cript_new_text = encode(ord_text, alphabet, key)
cript_new_text = np.array(cript_new_text)
cript_new_text = cript_new_text.reshape(-1)
print(cript_new_text)

# decode
ord_decode_new_text = decode(cript_new_text, alphabet, key)
ord_decode_new_text = np.array(ord_decode_new_text).reshape(-1)
print(ord_decode_new_text)
decode_new_text = list(map(lambda i: alphabet[i], ord_decode_new_text ))
# print  decode_new_text
print(decode_new_text)
```

```python
def get_stat(ord_text):
    """get statistics from baseline text"""
    # get list of counts
    #print("!!!!!", len(ord_text))
    ord_text = np.append(ord_text, [255])
    #print("!!!!!", len(ord_text))
    statistics = np.bincount(ord_text)
    #print(statistics)
    statistics[-1] -= 1
    
    #print(statistics)
    F = np.array(statistics)
    
    # sort for convinient
    alphabet_sort = [x for _,x in sorted(zip(F, alphabet))]
    F_sort = [x for x,_ in sorted(zip(F, alphabet))]

    #print(F_sort)
    #print(alphabet_sort)
    return np.array(F_sort), np.array(alphabet_sort)

def find_len_of_key(text, baseline_uc):
#     max_uc = 10000
    max_uc = 0
    max_uc_idx = 10000
    text_for_uc = []
    #print('len', len(text))
    for i in range(1, 20):
        ord_text_for_uc = [text[j] for j in range(0, len(text), i)]
        print('len ord_text_for_uc', len(ord_text_for_uc))
        F_sort, alphabet_sort = get_stat(ord_text_for_uc)
        # get uc
        uc = np.sum((F_sort / np.sum(F_sort)) * ((F_sort - 1) / np.sum(F_sort - 1)))
        print(uc)
        #if (baseline_uc - max_uc)**2 >= (baseline_uc - uc)**2:
        if (max_uc < uc):
            print("len", i)
            print("uc", uc)
            max_uc = uc
            max_uc_idx = i 
            max_alphabet_sort = alphabet_sort
    print(max_uc_idx)
    return max_uc_idx, max_uc, max_alphabet_sort

def find(string, ch):
    for idx, str_ch in enumerate(string):
        if ch == str_ch:
            return idx

def atack(text, ord_baseline_text, alphabet):
    
    print("stat")
    print(text)
    print(ord_baseline_text)
    
    F_sort, baseline_alphabet_sort = get_stat(ord_baseline_text)
    print(len(baseline_alphabet_sort))
    baseline_uc = F_sort / np.sum(F_sort) * ((F_sort - 1) / np.sum(F_sort - 1))
    print(" baseline uc", np.sum(baseline_uc))
    len_of_key, max_uc, max_alphabet_sort = find_len_of_key(text, np.sum(baseline_uc))
    print('len_of_key',len_of_key)
    key1 = '' 
    
    new_text = ['a'] * len(text) 
    
    for start in range(len_of_key):
        text_for_uc = [text[j] for j in range(start, len(text), len_of_key)]
        F_sort, alphabet_sort = get_stat(text_for_uc)
        print("sort alp", len(alphabet_sort))
        
        #uc = (F_sort / np.sum(F_sort)) * ((F_sort - 1) / np.sum(F_sort - 1))
        
        # take ch from baseline_alphabet_sort with the same idx in alphabet_sort

        key1 +=  alphabet[ord_alphabet[find(alphabet, alphabet_sort[-1])] - ord_alphabet[find(alphabet, baseline_alphabet_sort[-1])]]
    
    print("key:", len(key1), " " , key1)
    ord_key = [ord_alphabet[find(alphabet, ch)] for ch in key1]
    new_text = decode(text, alphabet, ord_key)
    return new_text

```
